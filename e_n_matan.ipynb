**Динамической системой** называется математическая модель, описывающая изменение состояния во времени

**Оператор эволюции** - правило, определяющее переход из текущего состояния в следующее

**Точечное отображение** - оператор эволюции для систем с дискретным временем, задаваемый формулой \( x_{n+1} = f(x_n) \)

**Дискретная динамическая система** задаётся рекуррентным соотношением:  

$$
\vec{x}_{n+1} = \vec{f}(\vec{x}_n),
$$

где $\vec{x}_n$  состояние в момент $n$, а $\vec{f}$ - оператор эволюции.  

**Логистическое отображение** - конкретное точечное отображение вида \( x_{n+1} = r x_n (1 - x_n) \), моделирующее рост популяции с ограниченными ресурсами.одномерное точечное отображение вида:

$$
x_{n+1} = r \cdot x_n \cdot (1 - x_n),
$$

где $x_n \in [0, 1]$, $r \in [0, 4]$.  

**Физический смысл:**  
 $x_n$  нормированная численность популяции;  
 $r$  коэффициент роста.


опечатка, должно быть (0,1]



## Теорема
Для логистического отображения  
$$
x_{n+1} = r x_n (1 - x_n), \quad r \in (0, 1], \quad 0 < x_0 < 1
$$  
выполняется:  
$$
0 < x_n < 1 \quad \text{для всех } n \in \mathbb{N}.
$$

---

## Доказательство

### 1. Анализ функции итерации
Рассмотрим $f(x) = r x (1 - x)$ на $[0, 1]$ при $r \in (0, 1]$.

 **Нули:** $f(0) = f(1) = 0$
 **Производная:** $f'(x) = r(1 - 2x)$
 **Максимум:** $f'(x) = 0 \Rightarrow x = \frac{1}{2}$
  $$
  f_{\max} = f\left(\frac{1}{2}\right) = \frac{r}{4} \leq \frac{1}{4} < 1
  $$

**Наблюдение:** При $x \in (0, 1)$ все множители $r$, $x$, $(1-x)$ положительны, значит $f(x) > 0$.

---

### 2. Индуктивное доказательство
**База:** $0 < x_0 < 1$ (дано).

**Предположение:** Пусть $0 < x_n < 1$ для некоторого $n$.

**Переход:**
1. **Нижняя граница:**  
   $x_{n+1} = r x_n (1 - x_n) > 0$ (все множители положительны).

2. **Верхняя граница:**  
   Поскольку $f(x) \leq f_{\max} < 1$ для всех $x \in [0, 1]$, то:
   $$
   x_{n+1} = f(x_n) \leq \frac{r}{4} \leq \frac{1}{4} < 1.
   $$

Следовательно, $0 < x_{n+1} < 1$.

По индукции утверждение верно для всех $n \in \mathbb{N}$.

---

### 3. Замечания
**Случай $r = 1$:** Все рассуждения сохраняются ($f_{\max} = 0.25 < 1$).

 **Случай $r = 0$:** Исключён как вырожденный ($x_{n+1} = 0$, нарушает условие при $n \geq 1$).

---

## Вывод
При $r \in (0, 1]$ траектория логистического отображения, начинающаяся в $(0, 1)$, никогда не покидает этот интервал. Это фундаментальное свойство, основанное на ограниченности максимума параболы $f(x)$.

## Влияние параметра $r$ на вид функции $f(x) = r x (1 - x)$

### Аналитический вывод

Функция $f(x) = r x (1 - x)$ представляет собой параболу с ветвями, направленными вниз. Её форма и свойства существенно зависят от параметра $r$:

1. **Положение нулей:**
   $f(0) = 0$ и $f(1) = 0$ для любого $r > 0$
   Нули фиксированы в точках $x = 0$ и $x = 1$, не зависят от $r$

2. **Максимум функции:**
   Координаты максимума: $x_{\text{max}} = \frac{1}{2}$ (не зависит от $r$)
   Значение максимума: $f_{\text{max}} = f\left(\frac{1}{2}\right) = \frac{r}{4}$
   **Ключевой вывод:** Высота параболы прямо пропорциональна $r$

3. **Пересечение с прямой $y = x$:**
   Уравнение $x = r x (1 - x)$ имеет решения:
      $x_1^* = 0$ (всегда)
      $x_2^* = 1 - \frac{1}{r}$ (при $r \geq 1$)
     При $r < 1$: только одна точка пересечения ($x = 0$)
     При $r > 1$: две точки пересечения
     При $r = 1$: $x_2^* = 0$ (точки сливаются)

### Классификация по диапазонам $r$

#### 1. $r \in (0, 1]$ - малый рост
Парабола целиком лежит **ниже** прямой $y = x$
Максимум: $f_{\text{max}} \leq 0.25$
Все итерации стремятся к $x = 0$
**Физический смысл:** Рождаемость недостаточна для поддержания популяции

#### 2. $r \in (1, 3]$ - умеренный рост
Парабола пересекает прямую $y = x$ в двух точках
Появляется ненулевая неподвижная точка $x^* = 1 - 1/r$
Формируется "воронка" сходимости к $x^*$
**Физический смысл:** Устанавливается равновесная популяция

#### 3. $r \in (3, 4]$ - большой рост
Крутизна параболы в точке пересечения $|f'(x^*)| > 1$
Появляются бифуркации, циклы, хаос
**Физический смысл:** Перенаселение → колебания численности

### Графическая иллюстрация

#### Рекомендуемые значения $r$ для построения:
1. **$r = 0.5$** - типичный представитель $(0, 1]$
2. **$r = 1.0$** - граничный случай
3. **$r = 2.0$** - пример $(1, 3)$ с устойчивой неподвижной точкой
4. **$r = 3.2$** - начало осцилляций (бифуркация удвоения периода)
5. **$r = 3.8$** - хаотический режим

### Визуальные особенности на графиках

1. **Высота параболы:**
   $$
   \text{Высота} \propto r
   $$
    $r = 0.5$: максимум 0.125
    $r = 2.0$: максимум 0.5
    $r = 4.0$: максимум 1.0

2. **"Ширина" параболы:**
    Кривизна увеличивается с ростом $r$
    При $r = 4$ парабола становится симметричной относительно $x = 0.5$

3. **Положение относительно $y = x$:**
    При $r < 1$: вся парабола ниже диагонали
    При $r > 1$: часть параболы выше диагонали

### Заключительный вывод

Параметр $r$ в логистическом отображении выполняет роль **управляющего параметра**, который определяет качественное поведение системы:

1. **Масштабирование** - линейно увеличивает амплитуду колебаний
2. **Изменение устойчивости** - переходы через критические значения $r = 1$, $r = 3$, $r ≈ 3.45$, etc.
3. **Контроль сложности** - от простой сходимости к нулю до детерминированного хаоса

График $f(x)$ при разных $r$ наглядно демонстрирует эту эволюцию: от низкой пологой параболы при малых $r$ до высокой, почти заполняющей единичный квадрат параболы при $r = 4$. Качественные изменения происходят именно в точках бифуркаций, где меняется число и устойчивость неподвижных точек.

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    """Логистическое отображение: x_{n+1} = r * x_n * (1 - x_n)"""
    return r * x * (1 - x)

# Настройка графики
plt.figure(figsize=(12, 8))
plt.rcParams.update({'font.size': 12})

# Значения r для демонстрации разных режимов
r_values = [0.5, 1.0, 2.0, 3.0, 3.5, 4.0]
colors = ['blue', 'green', 'orange', 'red', 'purple', 'brown']

# Создаём сетку x
x = np.linspace(0, 1, 500)

# Основной график
plt.plot(x, x, 'k--', linewidth=1.5, label='Диагональ: $x_{n+1} = x_n$')

# Рисуем параболы для каждого r
for r, color in zip(r_values, colors):
    y = logistic_map(x, r)
    plt.plot(x, y, linewidth=2.5, color=color, label=f'$r = {r}$')
    
    # Отмечаем максимум
    max_x = 0.5
    max_y = logistic_map(max_x, r)
    plt.plot(max_x, max_y, 'o', color=color, markersize=6)

# Настройки отображения
plt.xlabel('$x_n$ (текущее состояние)', fontsize=14)
plt.ylabel('$x_{n+1}$ (следующее состояние)', fontsize=14)
plt.title('Влияние параметра $r$ на форму логистического отображения', fontsize=16)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right')
plt.xlim(0, 1)
plt.ylim(0, 1.1)
plt.tight_layout()
plt.show()
```

# Анализ вариантного отображения (N=3)

## Функция отображения
$$
g(x) = r x (1 - x)(3 - x), \quad x \in [0, 1], \quad r \in [0, r_{\max}]
$$
где $r_{\max} \approx 1.585$.

---

## Свойства функции $g(x)$

### 1. Нули функции
Решаем $g(x) = 0$:
$$
r x (1 - x)(3 - x) = 0
$$
Получаем три нуля:
 $x = 0$ (всегда)
 $x = 1$ (из $(1 - x) = 0$)
 $x = 3$ (из $(3 - x) = 0$) **лежит вне [0,1]**

**Вывод:** На отрезке $[0, 1]$ функция имеет два нуля: $x=0$ и $x=1$.

---

### 2. Производная и экстремумы
$$
g'(x) = r \cdot \frac{d}{dx}[x(1-x)(3-x)]
$$


$$
x(1-x)(3-x) = x(3 - x - 3x + x^2) = x(3 - 4x + x^2) = 3x - 4x^2 + x^3
$$

Производная:
$$
g'(x) = r(3 - 8x + 3x^2)
$$

$$
3x^2 - 8x + 3 = 0
$$
$$
x_{1,2} = \frac{8 \pm \sqrt{64 - 36}}{6} = \frac{8 \pm \sqrt{28}}{6} = \frac{8 \pm 2\sqrt{7}}{6}
$$

$$
x_1 = \frac{4 - \sqrt{7}}{3} \approx 0.451, \quad
x_2 = \frac{4 + \sqrt{7}}{3} \approx 1.882
$$

Только $x_1 \approx 0.451$ лежит в $[0, 1]$.

---

### 3. Максимум функции
Значение в максимуме:
$$
g_{\max} = g(x_1) = r \cdot x_1 (1 - x_1)(3 - x_1)
$$


 $x_1 \approx 0.451$
 
 $(1 - x_1) \approx 0.549$
 
 $(3 - x_1) \approx 2.549$
 
 Произведение: $0.451 \times 0.549 \times 2.549 \approx 0.631$
 


$$
g_{\max} \approx 0.631r
$$

**Критерий $g_{\max} \leq 1$:**
$$
0.631r \leq 1 \quad \Rightarrow \quad r \leq \frac{1}{0.631} \approx 1.585
$$
Это согласуется с заданным $r_{\max}$.

---

## Сравнение с логистическим отображением

### Сходства:
1. **Форма:** Обе функции  полиномы 3-й степени (вариант) и 2-й степени (логистика)
2. **Нули:** Обе имеют нули в $x=0$ и $x=1$
3. **Ограниченность:** Максимальное значение ограничено сверху 1 при соответствующих $r$
4. **Физический смысл:** Моделируют рост с ограниченными ресурсами

### Различия:
| Параметр | Логистическое | Вариант N=3 |
|----------|---------------|-------------|
| Степень | 2 (парабола) | 3 (кубика) |
| Макс. положение | $x = 0.5$ | $x \approx 0.451$ |
| Форма максимума | Симметричный | Асимметричный |
| Коэффициент | $f_{\max} = r/4$ | $g_{\max} \approx 0.631r$ |
| Диапазон $r$ | $[0, 4]$ | $[0, 1.585]$ |

### Физическая интерпретация:
В варианте N=3 множитель $(3 - x)$ можно интерпретировать как:

Дополнительный фактор "благоприятности среды"

При $x \to 0$: множитель $\approx 3$  очень благоприятно

При $x \to 1$: множитель $\approx 2$  менее благоприятно
 
Более сложная модель взаимодействия популяции со средой

## Физические причины различий

### Множитель $(3-x)$ в варианте N=3
Этот множитель можно интерпретировать как:

1. **Фактор благоприятности среды:**
   - При $x=0$: множитель = 3  максимально благоприятно
   - При $x=1$: множитель = 2  менее благоприятно
   - Среда лучше поддерживает малые популяции

2. **Механизм саморегуляции:**
   - Не просто "1 - x" (линейное ограничение)
   - Добавлен дополнительный регулирующий фактор
   - Более сложная модель взаимодействия

3. **Асимметрия конкуренции:**
   - В логистике: конкуренция симметрична
   - В варианте: конкуренция усиливается с ростом популяции, но нелинейно

### Математическая структура
**Логистическое:** $rx(1-x)$  простейшая нелинейная модель  
**Вариант N=3:** $rx(1-x)(3-x)$  следующее приближение в ряду Тейлора

Можно рассматривать вариант как уточнение логистической модели:
$$
(3-x) = 3(1 - x/3) \approx \text{коррекция к } (1-x)
$$

Это делает модель более гибкой для описания реальных систем, но сужает диапазон управляющего параметра $r$, в котором система остаётся физически осмысленной (значения не выходят за [0, 1]).

## 1. Неподвижная точка 

**Определение:** Точка $x^*$ называется *неподвижной точкой* отображения $f$, если
$$
x^* = f(x^*)
$$

**Интерпретация:** Состояние системы, которое не меняется со временем. В контексте логистического отображения - численность популяции, остающаяся постоянной.

---

## 2. Подпоследовательность 

**Определение:** Пусть $\{x_n\}_{n=0}^\infty$ - последовательность. *Подпоследовательность*  это последовательность $\{x_{n_k}\}_{k=0}^\infty$, где $n_0 < n_1 < n_2 < \dots$  возрастающие индексы.

**Пример:** Чётные члены $\{x_{2k}\}$ и нечётные члены $\{x_{2k+1}\}$.

**Важность:** Позволяет анализировать сложное поведение, разбивая его на более простые компоненты.

---

## 3. Монотонность 

**Определение:** Последовательность $\{x_n\}$ называется:
- *Монотонно возрастающей*, если $x_{n+1} \geq x_n$ для всех $n$
- *Монотонно убывающей*, если $x_{n+1} \leq x_n$ для всех $n$
- *Строго монотонной*, если неравенства строгие

**Критерий для отображений:** Если $f(x) < x$ для всех $x$ в интервале, то последовательность монотонно убывает.

---

## 4. Сходимость 

**Определение:** Последовательность $\{x_n\}$ *сходится* к пределу $L$, если
$$
\forall \varepsilon > 0 \ \exists N: \ \forall n > N \ |x_n - L| < \varepsilon
$$
Обозначение: $\lim_{n \to \infty} x_n = L$.

**Связь с неподвижными точками:** Если последовательность сходится, то её предел неподвижная точка.

---

## 5. Предел последовательности 

**Определение:** Число $L$ называется *пределом* последовательности $\{x_n\}$, если члены последовательности неограниченно приближаются к $L$ при $n \to \infty$.

**Теорема (о пределе):** Если монотонная ограниченная последовательность, то она имеет предел.

---

## 6. Устойчивость

**Определение :** Неподвижная точка *устойчива*, если малые возмущения начального условия не приводят к уходу траектории далеко от точки.

**Критерий :** Точка $x^*$ устойчива, если $|f'(x^*)| < 1$.

---

### Определение
**Неподвижная точка** $x^*$ отображения $f$ удовлетворяет:
$$
x^* = f(x^*)
$$

Для логистического отображения $f(x) = r x (1 - x)$ уравнение имеет вид:
$$
x = r x (1 - x)
$$

### Решение


$$
x - r x (1 - x) = 0
$$


$$
x \left[1 - r(1 - x)\right] = 0
$$



**Случай : $x = 0$**
$$
x_1^* = 0
$$
Эта неподвижная точка существует при любом $r$.

**Случай : $x \neq 0$**
Тогда:
$$
1 - r(1 - x) = 0
$$
$$
1 - r + r x = 0
$$
$$
r x = r - 1
$$
$$
x_2^* = 1 - \frac{1}{r}
$$
Эта точка существует при $r \neq 0$.

### ответ:
**Неподвижные точки логистического отображения:**
$$
\boxed{x^* = 0 \quad \text{и} \quad x^* = 1 - \frac{1}{r} \ \text{(при } r \neq 0\text{)}}
$$

---




#### Уравнение неподвижных точек:
$$
x = r x (1 - x)
$$
После преобразования:
$$
x [1 - r(1 - x)] = 0
$$

#### Решения:
1. $x_1^* = 0$  всегда существует (при любом $r$)
2. $x_2^* = 1 - \frac{1}{r}$  существует при $r \neq 0$

---

### Случай 1: одна неподвижная точка

**Условие:** $x_1^*$ и $x_2^*$  это одна и та же точка, либо $x_2^*$ не входит в область определения $[0, 1]$.

#### Подслучаи:

** $r = 0$ - вырожденный случай**
- $x_1^* = 0$ существует
- $x_2^* = 1 - 1/0$  не определена
- **Итог:** 1 точка ($x^* = 0$)

** $0 < r < 1$  малый рост**
- $x_1^* = 0$ (в $[0, 1]$)
- $x_2^* = 1 - 1/r < 0$ (отрицательная, вне $[0, 1]$)
- **Итог:** 1 точка в $[0, 1]$ ($x^* = 0$)

** $r = 1$  критический случай**
- $x_1^* = 0$ 
- $x_2^* = 1 - 1/1 = 0$  совпадает с $x_1^*$
- **Итог:** 1 точка (две точки сливаются)

---

### Случай 2: несколько (две) неподвижных точек

**Условие:** Обе точки различны и принадлежат $[0, 1]$.

#### Единственный случай:

**$r > 1$  достаточный рост**
- $x_1^* = 0$ (в $[0, 1]$)
- $x_2^* = 1 - 1/r$:
  - При $r > 1$: $0 < 1 - 1/r < 1$
  - Пример: $r = 2$ → $x_2^* = 0.5$
  - Пример: $r = 3$ → $x_2^* ≈ 0.667$
  - Пример: $r = 4$ → $x_2^* = 0.75$
- **Итог:** 2 различные точки в $[0, 1]$

---

### Итоговая таблица

| Диапазон $r$ | $x_1^* = 0$ | $x_2^* = 1 - 1/r$ | Количество различных точек в $[0, 1]$ | Причина |
|--------------|-------------|-------------------|--------------------------------------|---------|
| **$r = 0$** |  |  (не определена) | **1** | Вырожденное отображение |
| **$0 < r < 1$** |  |  ($< 0$) | **1** | Вторая точка отрицательна |
| **$r = 1$** |  |  ($= 0$) | **1** | Точки совпадают |
| **$1 < r \leq 4$** |  |  ($0 < x_2^* < 1$) | **2** | Две различные точки |

---


```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
r_values = [0.5, 1.0, 2.0, 3.0, 4.0] # Значения r для демонстрации
colors = ['blue', 'green', 'orange', 'red', 'purple']

x = np.linspace(0, 1, 500)

for i, r in enumerate(r_values):
    plt.subplot(2, 3, i+1)

    y = r * x * (1 - x)    # Функция и диагональ
    plt.plot(x, y, 'b-', linewidth=2, label=f'$f(x)$, $r={r}$')
    plt.plot(x, x, 'k--', linewidth=1, label='$y=x$')
    fixed_points = []    # Неподвижные точки
    fixed_points.append(0)    # Всегда x=0
    plt.plot(0, 0, 'ro', markersize=8, label='$x^*=0$')
    
    if r >= 1:    # Вторая точка при r>=1
        x2 = 1 - 1/r
        if 0 <= x2 <= 1:
            fixed_points.append(x2)
            plt.plot(x2, x2, 'go', markersize=8, label=f'$x^*={x2:.2f}$')
    
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.title(f'$r = {r}$\nТочек: {len(fixed_points)}')
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=9)
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)

plt.suptitle('Неподвижные точки логистического отображения', fontsize=16)
plt.tight_layout()
plt.show()
```

Максимальное количество неподвижных точек

**Логистическое отображение может иметь не более двух неподвижных точек.**

---

### Доказательство 1: Алгебраический подход

Уравнение для неподвижных точек:
$$
x = r x (1 - x)
$$

Преобразуем:
1. Переносим всё в одну сторону:
   $$
   x - r x (1 - x) = 0
   $$

2. Раскрываем скобки:
   $$
   x - r x + r x^2 = 0
   $$

3. Группируем:
   $$
   r x^2 + (1 - r) x = 0
   $$

4. Выносим $x$:
   $$
   x \left[ r x + (1 - r) \right] = 0
   $$

**Получили:** Уравнение второй степени относительно $x$.

**Математический факт:** Квадратное уравнение имеет не более двух действительных корней.

---

### Доказательство 2: Геометрический подход

Неподвижные точки - это точки пересечения:
1. Графика функции: $y = f(x) = r x (1 - x)$ (парабола, ветви вниз)
2. Прямой: $y = x$ (диагональ)

**Анализ пересечений:**
- Парабола и прямая - кривые первого и второго порядка
- Максимальное число пересечений параболы и прямой  **2**

## Задача 2: Монотонность и сходимость при r ∈ (0, 1]

### Постановка задачи
Дано: 
- Логистическое отображение: $x_{n+1} = r x_n (1 - x_n)$
- Начальное условие: $x_0 \in (0, 1)$
- Параметр: $r \in (0, 1]$

Требуется:
1. Доказать, что последовательность $\{x_n\}$ монотонно убывает
2. Доказать существование предела
3. Показать графически

---

## Теорема о монотонности и сходимости при r ∈ (0, 1]

### Утверждение
Для логистического отображения $x_{n+1} = r x_n (1 - x_n)$ при $x_0 \in (0, 1)$ и $r \in (0, 1]$:
1. Последовательность $\{x_n\}$ **строго монотонно убывает**
2. Предел существует и равен **0**

---

### Доказательство

#### Часть 1: Монотонное убывание

**Шаг 1: Вспомогательные факты**
Из Easy level известно, что при данных условиях:
$$
0 < x_n < 1 \quad \text{для всех } n \in \mathbb{N}
$$

**Шаг 2: Ключевое наблюдение**
Рассмотрим отношение последовательных членов:
$$
\frac{x_{n+1}}{x_n} = \frac{r x_n (1 - x_n)}{x_n} = r(1 - x_n)
$$

**Шаг 3: Анализ функции $\varphi(x) = r(1 - x)$**

Для $x \in (0, 1)$ функция $\varphi(x)$ обладает свойствами:
1. Линейно убывает от $\varphi(0) = r$ до $\varphi(1) = 0$
2. При $r \in (0, 1]$:
   - Максимум: $\varphi_{\max} = \varphi(0) = r \leq 1$
   - Минимум: $\varphi_{\min} = \varphi(1) = 0$

**Шаг 4: Оценка отношения**

Для всех $x_n \in (0, 1)$:
1. **Если $0 < r < 1$:**
   $$ 
   \frac{x_{n+1}}{x_n} = r(1 - x_n) \leq r < 1
   $$

2. **Если $r = 1$:**
   $$
   \frac{x_{n+1}}{x_n} = 1 - x_n < 1 \quad (\text{так как } x_n > 0)
   $$

В обоих случаях:
$$
\frac{x_{n+1}}{x_n} < 1
$$

**Шаг 5: Заключение о монотонности**

Поскольку все $x_n > 0$, из $\frac{x_{n+1}}{x_n} < 1$ следует:
$$
x_{n+1} < x_n \quad \text{для всех } n
$$

Таким образом, последовательность $\{x_n\}$ **строго монотонно убывает**.

---

#### Часть 2: Существование и значение предела

**Шаг 6: Применение теоремы Вейерштрасса**

Последовательность обладает свойствами:
1. Ограничена снизу: $0 < x_n$ (из Easy level)
2. Монотонно убывает (доказано выше)

По **теореме Вейерштрасса** о пределе монотонной ограниченной последовательности:
$$
\exists L = \lim_{n \to \infty} x_n
$$

**Шаг 7: Нахождение предела**

Перейдём к пределу в уравнении отображения:
$$
\lim_{n \to \infty} x_{n+1} = \lim_{n \to \infty} [r x_n (1 - x_n)]
$$

Поскольку предел существует:
$$
L = r L (1 - L)
$$

Решаем уравнение неподвижных точек:
$$
L - r L (1 - L) = 0
$$
$$
L[1 - r(1 - L)] = 0
$$

Возможные решения:
$$
\begin{cases}
L_1 = 0 \\[4pt]
L_2 = 1 - \dfrac{1}{r} \quad (\text{при } r \neq 0)
\end{cases}
$$

**Шаг 8: Выбор физически осмысленного предела**

Анализируем $L_2$:
1. **При $0 < r < 1$:**
   $$ 
   L_2 = 1 - \frac{1}{r} < 0 
   $$
   Не может быть пределом положительной последовательности.

2. **При $r = 1$:**
   $$
   L_2 = 1 - \frac{1}{1} = 0 = L_1
   $$
   Решения совпадают.

Последовательность монотонно убывает, начиная с $x_0 > 0$. Если бы $L = L_2 > 0$, то последовательность должна была бы убывать к положительному числу, что возможно, но в данном случае $L_2$ либо отрицателен, либо равен 0.

**Шаг 9: Итоговый результат**
$$
\boxed{\lim_{n \to \infty} x_n = 0}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_trajectory(x0, r, n_iter=20):
    """Генерирует траекторию логистического отображения"""
    trajectory = [x0]
    x = x0
    for _ in range(n_iter):
        x = r * x * (1 - x)
        trajectory.append(x)
    return trajectory

# Параметры
x0 = 0.8  # Начальное значение
r_values = [0.3, 0.6, 0.9, 1.0]  # Разные значения r из (0, 1]
n_iter = 15

plt.figure(figsize=(14, 10))
for i, r in enumerate(r_values, 1):  # 1. Графики траекторий
    plt.subplot(2, 2, i)
    traj = logistic_trajectory(x0, r, n_iter)
    plt.plot(range(len(traj)), traj, 'bo-', linewidth=2, markersize=5,     # Рисуем траекторию
             label=f'$x_0={x0}$, $r={r}$')
    
    plt.axhline(y=0, color='r', linestyle='--', alpha=0.7,     # Горизонтальная линия на уровне 0
                linewidth=1.5, label='Предел: $x^*=0$')
    
    plt.xlabel('Номер итерации $n$', fontsize=11)
    plt.ylabel('Значение $x_n$', fontsize=11)
    plt.title(f'Монотонное убывание при $r = {r}$', fontsize=12)
    plt.grid(True, alpha=0.3, linestyle=':')
    plt.legend(fontsize=10)
    plt.ylim(-0.05, 0.85)
    
    last_value = traj[-1]      # информация о скорости убывания
    plt.text(n_iter//2, 0.7, f'$x_{{{n_iter}}} = {last_value:.4f}$', 
             fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

plt.suptitle(r'Монотонное убывание логистического отображения при $r \in (0, 1]$', 
             fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6)) # 2. Дополнительный график: сравнение скоростей сходимости

x0 = 0.8
n_iter = 25

for r in [0.3, 0.6, 0.9, 1.0]:
    traj = logistic_trajectory(x0, r, n_iter)
    plt.plot(range(len(traj)), traj, linewidth=2.5, 
             label=f'$r={r}$', marker='o', markersize=3, markevery=2)

plt.axhline(y=0, color='k', linestyle='--', alpha=0.5, label='Предел $x^*=0$')
plt.xlabel('Итерация $n$', fontsize=13)
plt.ylabel('$x_n$', fontsize=13)
plt.title('Сравнение скоростей сходимости к 0 при разных $r$', fontsize=14)
plt.grid(True, alpha=0.2)
plt.legend(fontsize=12)
plt.yscale('log')  # Логарифмическая шкала для наглядности
plt.ylim(1e-10, 1)
plt.tight_layout()
plt.show()
```

## Доказательство монотонности подпоследовательностей

### Дано:
- Логистическое отображение: $x_{n+1} = r x_n (1 - x_n)$
- $r \in (2, 3)$
- Неподвижная точка: $x^* = 1 - \frac{1}{r}$
- Условия: $x_{2n} > x^*$, $x_{2n+1} < x^*$

#### 1. Ключевое свойство
При $r > 2$:
- $x^* = 1 - 1/r > 0.5$
- На участке $x > 0.5$ функция $f(x) = r x (1 - x)$ **убывает**

Для убывающей функции: если $a > b$, то $f(a) < f(b)$.

#### 2. Осцилляция вокруг $x^*$
Из $x_{2n} > x^*$ и убывания $f(x)$:
$$
x_{2n+1} = f(x_{2n}) < f(x^*) = x^*
$$

Из $x_{2n+1} < x^*$:
$$
x_{2n+2} = f(x_{2n+1}) > f(x^*) = x^*
$$

**Вывод:** значения "прыгают" с одной стороны $x^*$ на другую.

#### 3. Монотонность чётных членов
Рассмотрим две итерации: $g(x) = f(f(x))$

Так как $f(x)$ убывает, то $g(x)$ возрастает (убывающая от убывающей).

Для $x > x^*$ график $y = g(x)$ лежит ниже диагонали $y = x$:
$$
g(x) < x \quad \text{при } x > x^*
$$

Применяем к $x_{2n} > x^*$:
$$
x_{2(n+1)} = g(x_{2n}) < x_{2n}
$$

⇒ $\{x_{2n}\}$ **монотонно убывает**.

#### 4. Монотонность нечётных членов
Для $x < x^*$ график $y = g(x)$ лежит выше диагонали:
$$
g(x) > x \quad \text{при } x < x^*
$$

Применяем к $x_{2n+1} < x^*$:
$$
x_{2(n+1)+1} = g(x_{2n+1}) > x_{2n+1}
$$

⇒ $\{x_{2n+1}\}$ **монотонно возрастает**.

---

```python
#лестница Ламерея 
import numpy as np
import matplotlib.pyplot as plt

def logistic(x, r):
    return r * x * (1 - x)
r = 2.8
fixed_point = 1 - 1/r
x0 = 0.8
n_steps = 10

x_vals = np.linspace(0, 1, 400)
y_vals = logistic(x_vals, r)

trajectory = [x0]
curr = x0
for _ in range(n_steps):
    curr = logistic(curr, r)
    trajectory.append(curr)

plt.figure(figsize=(10, 8)) # Лестница Ламерея

plt.plot(x_vals, y_vals, '#2980b9', linewidth=2.5, label='$f(x) = r x (1-x)$')
plt.plot(x_vals, x_vals, 'k--', alpha=0.5, linewidth=1.5, label='$y = x$')

plt.scatter([fixed_point], [fixed_point], color='green', s=120,  # Неподвижная точка
            edgecolor='black', linewidth=1.5, zorder=10,
            label=f'$x^* = {fixed_point:.3f}$')


path_x, path_y = [], [] # Построение лестницы
x_curr = x0
y_curr = 0

for i in range(n_steps):
    y_next = logistic(x_curr, r)     # Вертикаль: от y=x до параболы
    path_x.extend([x_curr, x_curr])
    path_y.extend([y_curr, y_next])
    path_x.extend([x_curr, y_next])    # Горизонталь: к диагонали
    path_y.extend([y_next, y_next])
    
    x_curr, y_curr = y_next, y_next

plt.plot(path_x, path_y, 'r-', alpha=0.7, linewidth=1.2, label='Траектория')

# Чётные и нечётные точки 
even_points = []  # x0, x2, x4...
odd_points = []   # x1, x3, x5...

for i, x_val in enumerate(trajectory):
    y_val = logistic(x_val, r) if i < len(trajectory)-1 else None
    
    if i % 2 == 0:  # Чётные
        even_points.append((x_val, y_val if y_val else fixed_point))
    else:           # Нечётные
        odd_points.append((x_val, y_val if y_val else fixed_point))

if even_points:
    even_x, even_y = zip(*even_points[:-1])  # Все кроме последнего
    plt.scatter(even_x, even_y, color='purple', marker='^', s=100,
                edgecolor='black', linewidth=1.5, zorder=10,
                label='Чётные $x_{2n}$')

if odd_points:
    odd_x, odd_y = zip(*odd_points[:-1])  # Все кроме последнего
    plt.scatter(odd_x, odd_y, color='orange', marker='v', s=100,
                edgecolor='black', linewidth=1.5, zorder=10,
                label='Нечётные $x_{2n+1}$')

plt.xlabel('$x_n$', fontsize=12)
plt.ylabel('$x_{n+1}$', fontsize=12)
plt.title(f'Лестница Ламерея при $r = {r}$\nЧётные точки выше $x^*$, нечётные ниже $x^*$', fontsize=14)
plt.grid(True, alpha=0.2, linestyle=':')
plt.legend(loc='upper right', fontsize=11)

# Автоматические пределы, чтобы всё поместилось
all_x = list(x_vals) + [x for x, _ in even_points] + [x for x, _ in odd_points] + [fixed_point]
all_y = list(y_vals) + [y for _, y in even_points] + [y for _, y in odd_points] + [fixed_point]

x_min, x_max = min(all_x), max(all_x)
y_min, y_max = min(all_y), max(all_y)

margin = 0.05
plt.xlim(x_min - margin, x_max + margin)
plt.ylim(y_min - margin, y_max + margin)

plt.tight_layout()
plt.show()
```

## Анализ варианта N=3: $g(x) = r x (1 - x)(3 - x)$

### 1. Аналитическое нахождение неподвижных точек

#### Уравнение неподвижной точки:
$$
x = g(x) = r x (1 - x)(3 - x)
$$

#### Решение:


$$
x - r x (1 - x)(3 - x) = 0
$$


$$
x \left[ 1 - r (1 - x)(3 - x) \right] = 0
$$

Первая неподвижная точка:
$$
x_1^* = 0
$$
Эта точка существует при любом $r$.

Вторая неподвижная точка (при $x \neq 0$):
$$
1 - r (1 - x)(3 - x) = 0
$$
$$
r (1 - x)(3 - x) = 1
$$

$$
r (3 - x - 3x + x^2) = 1
$$
$$
r (3 - 4x + x^2) = 1
$$
$$
r x^2 - 4r x + 3r - 1 = 0
$$

Это квадратное уравнение относительно $x$.


$$
D = (-4r)^2 - 4r(3r - 1) = 16r^2 - 12r^2 + 4r = 4r^2 + 4r = 4r(r + 1)
$$

Корни квадратного уравнения:
$$
x_{2,3}^* = \frac{4r \pm \sqrt{4r(r+1)}}{2r} = \frac{4r \pm 2\sqrt{r(r+1)}}{2r} = 2 \pm \sqrt{1 + \frac{1}{r}}
$$

---

#### Анализ корней для $r \in [0, r_{\max}]$, $r_{\max} \approx 1.585$:

1. **Корень $x_2^* = 2 - \sqrt{1 + \frac{1}{r}}$:**
   - При $r > 0$: $\sqrt{1 + \frac{1}{r}} > 1$, значит $x_2^* < 1$
   - При $r \to 0^+$: $x_2^* \to 2 - \infty = -\infty$
   - При $r \to \infty$: $x_2^* \to 2 - 1 = 1$
   - Нас интересует $x \in [0, 1]$

2. **Корень $x_3^* = 2 + \sqrt{1 + \frac{1}{r}}$:**
   - Всегда $> 2$, не лежит в $[0, 1]$

3. **Критическое значение $r$ для вхождения $x_2^*$ в $[0, 1]$:**
   Решаем $x_2^* = 0$:
   $$
   2 - \sqrt{1 + \frac{1}{r}} = 0
   $$
   $$
   \sqrt{1 + \frac{1}{r}} = 2
   $$
   $$
   1 + \frac{1}{r} = 4
   $$
   $$
   \frac{1}{r} = 3
   $$
   $$
   r = \frac{1}{3}
   $$

---

#### Итог по неподвижным точкам:

| Диапазон $r$ | Неподвижные точки в $[0, 1]$ | Комментарий |
|--------------|-----------------------------|-------------|
| **$r = 0$** | $x^* = 0$ | Вырожденный случай |
| **$0 < r < \frac{1}{3}$** | $x_1^* = 0$ | $x_2^* < 0$ (вне области) |
| **$r = \frac{1}{3}$** | $x_1^* = 0$ | $x_2^* = 0$ (сливаются) |
| **$\frac{1}{3} < r \leq r_{\max}$** | $x_1^* = 0$, $x_2^* = 2 - \sqrt{1 + 1/r} \in (0, 1)$ | Две точки |

---

### 2. Диапазон $r$ для монотонной сходимости к 0

#### Критерий монотонной сходимости к 0:
1. $0 < g(x) < x$ для всех $x \in (0, 1)$
2. $g'(0) < 1$ (условие устойчивости точки $x=0$)

#### Вычисление производной в 0:
$$
g'(x) = r \cdot \frac{d}{dx}[x(1-x)(3-x)]
$$

Раскроем произведение:
$$
x(1-x)(3-x) = x(3 - x - 3x + x^2) = x(3 - 4x + x^2) = 3x - 4x^2 + x^3
$$

Производная:
$$
g'(x) = r(3 - 8x + 3x^2)
$$

В точке $x=0$:
$$
g'(0) = 3r
$$

#### Условие устойчивости $x=0$:
$$
|g'(0)| < 1 \quad \Rightarrow \quad 3r < 1 \quad \Rightarrow \quad r < \frac{1}{3}
$$

#### Проверка условия $g(x) < x$ для $x \in (0, 1)$:
Рассмотрим отношение:
$$
\frac{g(x)}{x} = r(1-x)(3-x)
$$

На $[0, 1]$ функция $\varphi(x) = (1-x)(3-x)$:
- Максимум при $x=0$: $\varphi(0) = 3$
- Минимум при $x=1$: $\varphi(1) = 0$

Для выполнения $g(x) < x$ нужно:
$$
r(1-x)(3-x) < 1 \quad \text{для всех } x \in (0, 1)
$$

Достаточное условие: $r \cdot \max_{x \in [0,1]} \varphi(x) < 1$ ⇒ $3r < 1$ ⇒ $r < \frac{1}{3}$

---

#### Итог для пункта 2:

**Диапазон $r$ для монотонной сходимости к 0:**
$$
\boxed{0 < r < \frac{1}{3}}
$$

**Объяснение:**
1. При $r < 1/3$: $g'(0) = 3r < 1$ ⇒ точка $x=0$ устойчива
2. При $r < 1/3$: $g(x) < x$ для всех $x \in (0, 1)$ ⇒ монотонное убывание
3. При $r = 1/3$: граничный случай (касание)
4. При $r > 1/3$: появляется вторая неподвижная точка, сходимость к 0 нарушается


```python
import numpy as np
import matplotlib.pyplot as plt

def variant_map(x, r):
    """Вариант N=3: g(x) = r * x * (1 - x) * (3 - x)"""
    return r * x * (1 - x) * (3 - x)

def generate_trajectory(x0, r, n_iter):
    """Генерирует траекторию"""
    traj = [x0]
    x = x0
    for _ in range(n_iter):
        x = variant_map(x, r)
        traj.append(x)
    return traj

x0 = 0.8
n_iter = 20
r_values = [0.2, 0.4, 0.6, 1.0, 1.5]  # Разные r, включая критические
r_max = 27 / (2 * (7 * np.sqrt(7) - 10))  # ≈ 1.585

fig, axes = plt.subplots(len(r_values), 1, figsize=(10, 4*len(r_values)))

if len(r_values) == 1: # Если только один график, axes не будет списком
    axes = [axes]

for i, (ax, r) in enumerate(zip(axes, r_values)):
    if r > r_max:     # Проверка, что r в допустимом диапазоне
        ax.text(0.5, 0.5, rf'$r = {r} > r_{{\max}} \approx {r_max:.3f}$', 
                ha='center', va='center', transform=ax.transAxes, fontsize=12)
        ax.set_title(f'$r = {r}$ (вне допустимого диапазона)', fontsize=12)
        ax.set_xticks([])
        ax.set_yticks([])
        continue
    
    traj = generate_trajectory(x0, r, n_iter)
    ax.plot(range(len(traj)), traj, 'b-o', linewidth=2, markersize=5, 
            label=f'$x_0={x0}$')
    
# Неподвижные точки
    ax.axhline(y=0, color='r', linestyle='--', alpha=0.7, linewidth=1.5, 
               label='$x^*=0$')     # Всегда x=0
    if r > 1/3:     # Вторая точка при r > 1/3
        x2 = 2 - np.sqrt(1 + 1/r)
        if 0 <= x2 <= 1:
            ax.axhline(y=x2, color='g', linestyle='--', alpha=0.7, linewidth=1.5,
                      label=f'$x^*={x2:.3f}$')
    
    ax.set_xlabel('Итерация $n$', fontsize=11)
    ax.set_ylabel('$x_n$', fontsize=11)

    if r < 1/3:      # Определяем тип поведения
        behavior = "Монотонная сходимость к 0"
        color = 'green'
    elif r == 1/3:
        behavior = "Граничный случай ($r=1/3$)"
        color = 'orange'
    else:  # r > 1/3
        x2 = 2 - np.sqrt(1 + 1/r)
        behavior = f"Сходимость к $x^*={x2:.3f}$"
        color = 'red'
    
    ax.set_title(f'$r = {r}$: {behavior}', fontsize=12, color=color, pad=10)
    ax.grid(True, alpha=0.3, linestyle=':')
    ax.legend(fontsize=10, loc='upper right')
    
    y_max = max(traj) * 1.1        # Диапазон по Y
    ax.set_ylim(-0.05, min(1.1, y_max))
    
    ax.set_xticks(range(0, len(traj), 2))     # Добавляем сетку по X для лучшей читаемости

plt.suptitle(r'Поведение варианта N=3: $g(x) = r x (1 - x)(3 - x)$', 
             fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
